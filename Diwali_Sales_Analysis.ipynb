```python
sns.countplot(x='Gender', data = df)
```

```python
ax = sns.countplot(x='Gender', data = df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
ax = sns.countplot(x='Age Group',data = df, hue = 'Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by='Amount',ascending = False)
sns.barplot(x='Age Group', y= 'Amount', data= sales_age)
```

```python
ax = sns.countplot(x='State',data = df)
```

```python
sales_state = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=sales_state)
```

```python
sales_order = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(15,7)})
ax=sns.barplot(x='State',y='Orders',data=sales_order)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.columns
```

```python
sales_occ = df.groupby(['Occupation','Gender'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(15,7)})
ax=sns.barplot(x='Occupation',y='Orders',data=sales_occ,hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_product = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,10)})
sns.barplot(x='Product_Category',y='Amount',data=sales_product)
#for bars in ax.containers:
 #ax.bar_label(bars)
```

```python
sales_product = df.groupby(['Product_Category'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(x='Product_Category',y='Orders',data=sales_product)
```
